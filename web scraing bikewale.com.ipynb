```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.bikewale.com/bajaj-bikes/'
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text ,'lxml')
```

```python
soup
```

```python
bikes_name = []
engine = []
average = []
power_bhp = []
prices = []
bike_name = []
```

```python
bike = soup.find_all("h3" ,class_ = "o-jjpuv o-cVMLxW o-mHabQ o-fzpibK")
```

```python
bike
```

```python
for i in bike:
    bk = i.text
    bikes_name.append(bk)
```

```python
print(bikes_name)
print(len(bikes_name))
```

```python
pulsar_bikes_name= bikes_name[:-1]
pulsar_bikes_name
```

```python
bike_price = soup.find_all("span" ,class_ = "o-eZTujG o-byFsZJ o-bkmzIL o-bVSleT")
bike_price
```

```python
value = []
for j in bike_price:
    bp = j.text
    value.append(bp)
```

```python
print(value)
print(len(value))
```

```python
descriptions = soup.find_all("div" ,class_ = "o-bqHweY o-bVSleT o-bwCunT o-bfyaNx o-bNxxEB o-fzpihx")
```

```python
description = []
for j in descriptions:
    desc = j.text
    description.append(desc)
```

```python
description
```

```python
print(len(description))
```

```python
df = pd.DataFrame({'Bike_name':pulsar_bikes_name , 'Price_Exshowroom':value ,'Description':description})
```

```python
df
```

```python
df[['engine_capacity', 'Mileage' ,'power' ,'weight']] = df['Description'].str.split(pat = '|', expand = True)
```

```python
df.drop('Description', axis = 1, inplace = True)
```
